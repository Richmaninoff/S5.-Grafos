## Ejercicio 1
- Plot a colored version of the nodes based on attributes of the previous plot.
- Plot using Pyvis.
- Export the network: nx.write_gml(EliteNet, os.path.join("graphdata","EliteNet.gml"))
- Open the exported network in R and Gephi. Plot the graphs.

```python
node_attributes = nx.get_node_attributes(EliteNet, "multi")

#diccionario para almacenar los colores
colores = {}

for nodo, valor in node_attributes.items():
    # Verifica el valor del atributo 'multi' y asigna un color ('red' o 'green') al nodo
    if valor == 1:
        colores[nodo] = 'blue'
    else:
        colores[nodo] = 'red'
        
#existe un color llamado peru 
```

```python
nx.set_node_attributes(EliteNet, colores, "color")
```

```python
node_colors = [EliteNet.nodes[node]['color'] for node in EliteNet.nodes]
```

```python
nx.draw_random(EliteNet, node_color=node_colors, edge_color='silver', with_labels=True, font_size=8)
plt.show()
```
